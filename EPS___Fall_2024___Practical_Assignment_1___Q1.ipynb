```r

lambda <- 4  # rate of arrival per hour
n_simulations <- 10000  # number of simulations
delta_t <- 0.01  # small time step in minutes
time_since_last_customer <- 10 # 10 minutes have already passed

simulate_next_customer <- function(lambda, delta_t, time_since_last_customer) {
    lambda_per_min <- lambda / 60
    current_time <- time_since_last_customer

    while(TRUE){
        p_arrival <- lambda_per_min * delta_t

        if (runif(1) < p_arrival){
            return(current_time - time_since_last_customer)
        }

        current_time <- current_time + delta_t
    }
}

results <- replicate(n_simulations, simulate_next_customer(lambda, delta_t, time_since_last_customer))
hist(results,
     breaks=100,
     main="Number of Customer in Each Minute (10 min since last customer)",
     xlab="Time until next customer (minutes)",
     ylab="Frequency")
```

```r
lambda <- 4  # rate of arrival per hour
n_simulations <- 10000  # number of simulations
delta_t <- 0.01  # small time step in minutes
time_since_last_customer <- 40  # 40 minutes have already passed

simulate_next_three_customers <- function(lambda, delta_t, time_since_last_customer) {
    lambda_per_min <- lambda / 60
    current_time <- time_since_last_customer
    sum_of_times <- 0
    counter <- 0

    while(TRUE){
        p_arrival <- lambda_per_min * delta_t

        if (runif(1) < p_arrival){
            sum_of_times <- sum_of_times + (current_time - time_since_last_customer)
            counter <- counter + 1
            if(counter > 2){
                return(sum_of_times)
            }
        }

        current_time <- current_time + delta_t
    }
}

results <- replicate(n_simulations, simulate_next_three_customers(lambda, delta_t, time_since_last_customer))
hist(results,
     breaks=100,
     main="Number of Customer in Each Minute (10 min since last customer)",
     xlab="Time until next customer (minutes)",
     ylab="Frequency")
```
