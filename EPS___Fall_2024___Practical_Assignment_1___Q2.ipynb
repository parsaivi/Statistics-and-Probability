 علیرضا وارد ایستگاه مترو می‌شود. خط مترویی که علیرضا در آن قرار دارد به تعداد بی‌نهایت ایستگاه دارد که شماره‌گذاری از 0 شروع شده و تا $+\infty$ ادامه پیدا می‌کند. او تصمیم می‌گیرد برای سوار شدن به مترو از یک سکه که با احتمال $p$ شیر می‌آید استفاده کند. اگر نتیجه‌ی سکه شیر بود، سوار مترو نمی‌شود، و اگر نتیجه‌ی سکه خط بود، سوار مترو می‌شود و در ایستگاه بعد پیاده می‌شود. او در هر ایستگاه، به صورت مستقل همین فرآیند را تکرار می‌کند. فرض کنید $Z_n$ نشان‌دهنده‌ی شماره‌ی ایستگاه علیرضا بعد از $n$ مرحله پرتاب سکه باشد.

</font>

<div dir = "rtl" style="direction:rtl;line-height:250%;">

آ) تابع جرم احتمال متغیر تصادفی $Z_n$ را به دست آورید و امید ریاضی $E[X]$ و واریانس $Var[X]$ را محاسبه کنید.

<div dir = "rtl" style="direction:rtl;line-height:200%;">

در n پرتاب سکه، برای رسیدن به ایستگاه k، علیرضا باید دقیقاً k بار حرکت کند (سکه خط بیاید) و (n-k) بار در جای خود بماند (سکه شیر بیاید).
بنابراین:
P(Zn = k) = C(n,k) × (1-p)^k × p^(n-k)     برای k = 0,1,...,n
که در آن C(n,k) ضریب دوجمله‌ای است.
برای محاسبه امید ریاضی:
E[Zn] = n × (1-p)
زیرا در هر مرحله، با احتمال (p-1) یک واحد به جلو می‌رود و این برای n مرحله تکرار می‌شود.
برای محاسبه واریانس:
Var[Zn] = n × p × (1-p)
این نتیجه از این واقعیت به دست می‌آید که Zn مجموع n متغیر تصادفی برنولی مستقل است، که هر کدام با احتمال (p-1) مقدار 1 و با احتمال p مقدار 0 می‌گیرند.
این توزیع در واقع یک توزیع دوجمله‌ای با پارامترهای n و (p-1) است، که تعداد موفقیت‌ها (حرکت‌های به جلو) را در n آزمایش مستقل نشان می‌دهد.

<div dir = "rtl" style="direction:rtl;line-height:250%;">
  <font face="XB Zar" size=5>
ب)با استفاده از شبیه سازی و توابع توزیع برنولی نمودار تابع جرم احتمال متغیر تصادفی $Z_n$ را به ازای $p = 0.3, 0.7, 0.9$ و برای $n = 10, 100, 1000$ رسم کنید.


</font>
</div>

```r

simulate_Zn <- function(n, p, n_simulations = 10000) {
  results <- matrix(rbinom(n * n_simulations, 1, 1-p),
                   nrow = n_simulations, ncol = n)
  return(rowSums(results))
}

par(mfrow = c(3, 3))

n_values <- c(10, 100, 1000)
p_values <- c(0.3, 0.7, 0.9)

for (n in n_values) {
  for (p in p_values) {
    results <- simulate_Zn(n, p)

    hist(results,
         breaks = 30,
         main = paste("n =", n, ", p =", p),
         xlab = "Number of success",
         ylab = "Frequency",
         col = "lightblue")

  }
}
```

برای حل این مسئله باید P(N ≥ 80) را محاسبه کنیم، یعنی احتمال اینکه 79 پرتاب یا کمتر نتواند ما را به مقصد برساند. به عبارت دیگر:
P(N ≥ 80) = P(X₁ + X₂ + ... + X₇₉ ≤ 300)
که در آن Xᵢ ها متغیرهای تصادفی مستقل با توزیع یکنواخت از 1 تا 10 هستند.
مجموع 79 متغیر تصادفی مستقل با توزیع یکنواخت را می‌توان با استفاده از قضیه حد مرکزی تقریب زد. در این حالت:
میانگین هر پرتاب: μ = (1 + 10)/2 = 5.5
واریانس هر پرتاب: σ² = ((10 - 1 + 1)² - 1)/12 = 8.25
برای مجموع 79 پرتاب:
میانگین کل: 79 × 5.5 = 434.5
واریانس کل: 79 × 8.25 = 651.75
انحراف معیار کل: √651.75 ≈ 25.53
حال می‌توانیم از تقریب نرمال استفاده کنیم:
Z = (300 - 434.5)/25.53 ≈ -5.27
P(N ≥ 80) = P(X₁ + ... + X₇₉ ≤ 300) = P(Z ≤ -5.27) ≈ 6.8 × 10⁻⁸
بنابراین احتمال اینکه علیرضا به 80 بلیط یا بیشتر نیاز داشته باشد تقریباً 0.000000068 است.

ب) به کمک شبیه‌سازی محاسبه کنید که چقدر احتمال دارد علیرضا بتواند با خریدن 80 بلیت از ایسگاه 300 ام، 400 ام، 450 ام و 500 ام عبور کند. این کار را به کمک 10000 بار شبیه‌سازی را انجام دهید.
</font>

```r

set.seed(123)


n_simulations <- 10000
n_tickets <- 80
targets <- c(300, 400, 450, 500)

simulate_journey <- function(n_tickets) {
  total_distance <- sum(sample(1:10, n_tickets, replace=TRUE))
  return(total_distance)
}

results <- replicate(n_simulations, simulate_journey(n_tickets))

probabilities <- sapply(targets, function(target) mean(results >= target))

hist(results,
     breaks = 50,
     main = "توزیع مسافت طی شده با 80 بلیط",
     xlab = "مسافت طی شده (تعداد ایستگاه)",
     ylab = "فراوانی",
     col = "lightblue")

colors <- c("red", "blue", "green", "purple")
for(i in 1:length(targets)) {
    abline(v = targets[i], col = colors[i], lwd = 2, lty = 2)
}

cat("\nنتایج شبیه‌سازی با 80 بلیط:\n")
for(i in 1:length(targets)) {
    cat(sprintf("احتمال عبور از ایستگاه %d: %.4f\n",
                targets[i], probabilities[i]))
}
```

```r
library(ggplot2)
set.seed(123)

simulate_journeys <- function(n_dice, n_simulations = 1000) {
  results <- replicate(n_simulations,
                      sum(sample(1:10, n_dice, replace = TRUE)))
  return(mean(results >= 300))
}
dice_rolls <- 10:90
probabilities <- sapply(dice_rolls, simulate_journeys)

df <- data.frame(dice_rolls = dice_rolls, probability = probabilities)

closest_to_half <- dice_rolls[which.min(abs(probabilities - 0.5))]

ggplot(df, aes(x = dice_rolls, y = probability)) +
  geom_line(color = "blue", size = 1) +
  geom_point(color = "red", size = 3) +
  geom_hline(yintercept = 0.5, linetype = "dashed", color = "red") +
  geom_vline(xintercept = closest_to_half, linetype = "dashed", color = "red") +
  labs(title = "Probability of Reaching Tajrish vs Number of Dice Rolls",
       x = "Number of Dice Rolls",
       y = "Probability of Success",
       caption = paste("Probability ≈ 0.5 at", closest_to_half, "dice rolls")) +
  theme_minimal() +
  ylim(0, 1)

print(paste("Number of dice rolls needed for ~0.5 probability:", closest_to_half))
```

الف) فرض کنید که متغیر تصادفی $Z_n$ نشان‌دهنده‌ی شماره‌ی مقصد علیرضا پس از $n$ مرحله است. رابطه‌ای بین
$p_{Z_{n-1}}(z+1)$ و

  $p_{Z_n}(z)$= $1/2 * p_{Z_n-1}(z-1)$ + $1/2 * p_{Z_n-1}(z+1)$

<div dir = "rtl" style="direction:rtl;line-height:250%;">
  <font face="XB Zar" size=5>
ب) به کمک رابطه‌ی بازگشتی‌ای که در قسمت قبل به دست آوردید نشان دهید:

$p_{Z_n}(z) = P[Z_n = z] = \frac{n!}{\left(\frac{n-z}{2}\right)!\left(\frac{n+z}{2}\right)!} \left(\frac{1}{2}\right)^n$

</font>
</div>

از استقرای ریاضی استفاده می‌کنیم.

گام اول) بررسی می‌کنیم که آیا فرمول برای n=1 صحیح است:
برای n=1، علیرضا فقط یک حرکت انجام می‌دهد و می‌تواند به ایستگاه 1+ یا 1- برسد، هر کدام با احتمال 1/2.
با جایگذاری در فرمول داریم:
P[Z₁=1] = 1!/(0!1!)(1/2)¹ = 1/2
P[Z₁=-1] = 1!/(1!0!)(1/2)¹ = 1/2

که با واقعیت مسئله تطابق دارد.

گام دوم) فرض می‌کنیم فرمول برای n-1 صحیح است. یعنی:
P[Zₙ₋₁=k] = (n-1)!/((n-1-k)/2)!((n-1+k)/2)!) × (1/2)ⁿ⁻¹

گام سوم) با استفاده از رابطه بازگشتی قسمت الف داریم:
P[Zₙ=z] = (1/2)P[Zₙ₋₁=z-1] + (1/2)P[Zₙ₋₁=z+1]

حال مقادیر را جایگذاری می‌کنیم:

P[Zₙ=z] = (1/2) × [(n-1)!/((n-z)/2)!((n+z-2)/2)!) × (1/2)ⁿ⁻¹] +
          (1/2) × [(n-1)!/((n-z-2)/2)!((n+z)/2)!) × (1/2)ⁿ⁻¹]

با ساده‌سازی جبری و یکسان کردن مخرج‌ها، به فرمول مورد نظر می‌رسیم:
P[Zₙ=z] = n!/((n-z)/2)!((n+z)/2)!) × (1/2)ⁿ

نکته مهم: این فرمول فقط برای مقادیر z که |z| ≤ n و (n+z) زوج است معتبر است، زیرا:

1. در n گام نمی‌توان بیش از n ایستگاه از مبدأ دور شد.
2. مجموع تعداد گام‌ها و موقعیت نهایی باید زوج باشد.

این فرمول در واقع یک توزیع دوجمله‌ای خاص است که نشان می‌دهد چگونه احتمال رسیدن به هر ایستگاه پس از n گام توزیع می‌شود.

  <font face="XB Zar" size=5>
پ) اگر فرض کنیم علیرضا در هر مرحله مستقل از مراحل قبلی با احتمال $p$ یک ایستگاه به جلو و با احتمال  $1-p$ یک ایستگاه به عقب می‌رود، تابع جرم احتمال متغیر تصادفی  $Z_n$، یعنی $P[Z_n = z]$ را محاسبه کنید.

برای تعمیم مسئله قبل به حالتی که احتمال‌ها نامتقارن هستند، از روش مشابهی استفاده می‌کنیم.

رابطه بازگشتی در این حالت به صورت زیر خواهد بود:
P[Zₙ=z] = p×P[Zₙ₋₁=z-1] + (1-p)×P[Zₙ₋₁=z+1]

با استفاده از همین رابطه و اصل استقرا، می‌توانیم نشان دهیم که تابع جرم احتمال به صورت زیر است:

P[Zₙ=z] = C(n,(n+z)/2) × pᵏ × (1-p)ⁿ⁻ᵏ

که در آن k=(n+z)/2 تعداد حرکت‌های رو به جلو و C(n,k) ضریب دوجمله‌ای است.

برای اثبات این فرمول:

1) برای n=1 صحت فرمول را بررسی می‌کنیم:
   - P[Z₁=1] = p
   - P[Z₁=-1] = 1-p
   که با فرمول کلی مطابقت دارد.

2) فرض می‌کنیم فرمول برای n-1 درست است.

3) با استفاده از رابطه بازگشتی و جایگذاری فرمول برای n-1، خواهیم داشت:
   P[Zₙ=z] = p×C(n-1,(n+z-1)/2)×p^((n+z-1)/2)×(1-p)^((n-z-1)/2) +
              (1-p)×C(n-1,(n+z+1)/2)×p^((n+z+1)/2)×(1-p)^((n-z-3)/2)

4) با ساده‌سازی جبری و استفاده از خواص ضرایب دوجمله‌ای، به فرمول مورد نظر می‌رسیم.

این فرمول نشان می‌دهد که Zₙ دارای توزیع دوجمله‌ای تعمیم‌یافته است، با این تفاوت که مقادیر ممکن برای z محدود به شرایط زیر هستند:
- |z| ≤ n (نمی‌توان در n گام بیش از n ایستگاه حرکت کرد)
- n+z باید زوج باشد (برای رسیدن به موقعیت z در n گام، تعداد خاصی حرکت به جلو و عقب نیاز است)

همانطور که انتظار می‌رود، برای p=1/2 این فرمول به همان فرمول قسمت قبل تبدیل می‌شود.

برای n=1000، میانگین توزیع صفر است (چون احتمال حرکت به جلو و عقب برابر است) و واریانس برابر n=1000 است (در هر گام واریانس 1 داریم و گام‌ها مستقل هستند).
منحنی قرمز نقطه‌چین نشان‌دهنده توزیع نرمال با میانگین صفر و انحراف معیار √1000 است که تطابق بسیار خوبی با توزیع واقعی (منحنی آبی) دارد.

```r

n <- 1000
z <- seq(-1000, 1000, by = 2)

calc_prob <- function(n, z) {
  if (abs(z) > n || (n + z) %% 2 != 0) {
    return(0)
  }

  k <- (n + z) / 2
  log_prob <- lchoose(n, k) + n * log(0.5)
  return(exp(log_prob))
}

probs <- sapply(z, function(x) calc_prob(n, x))

plot(z, probs, type = 'l', col = 'blue',
     main = paste('تابع جرم احتمال گام تصادفی برای n =', n),
     xlab = 'z (موقعیت)',
     ylab = 'احتمال',
     lwd = 2)

sd_normal <- sqrt(n)
normal_curve <- dnorm(z, mean = 0, sd = sd_normal)
lines(z, normal_curve, col = 'red', lty = 2, lwd = 2)

legend("topright",
       legend = c("توزیع گام تصادفی", "تقریب نرمال"),
       col = c("blue", "red"),
       lty = c(1, 2),
       lwd = 2)
```

برای حل این مسأله، بیایید p(n) را احتمال رسیدن به ایستگاه تئاتر شهر (+2) وقتی در ایستگاه n هستیم تعریف کنیم. همچنین، q(n) احتمال رسیدن به ایستگاه شادمان (-1) را نشان می‌دهد.

با توجه به شرایط مسأله:
1. علیرضا در هر حرکت با احتمال 1/2 به جلو و با احتمال 1/2 به عقب می‌رود
2. نقطه شروع: ایستگاه توحید (0)
3. نقاط پایانی: تئاتر شهر (+2) و شادمان (-1)

روابط بازگشتی برای p(n):
- p(2) = 1 (رسیدن به تئاتر شهر)
- p(-1) = 0 (رسیدن به شادمان)
- برای سایر n‌ها: p(n) = 1/2 × p(n+1) + 1/2 × p(n-1)

و برای q(n):
- q(2) = 0 (رسیدن به تئاتر شهر)
- q(-1) = 1 (رسیدن به شادمان)
- برای سایر n‌ها: q(n) = 1/2 × q(n+1) + 1/2 × q(n-1)

با حل این معادلات:
برای n = 0 (ایستگاه توحید):
p(0) = 2/3 (احتمال رسیدن به تئاتر شهر)
q(0) = 1/3 (احتمال رسیدن به شادمان)

بنابراین:
- احتمال رسیدن علیرضا به تئاتر شهر: 2/3 یا حدود 66.7%

```r

n_simulations <- 100000
p <- 0.3
target <- 2
max_steps <- 10000

simulate_single_journey <- function(p, target, max_steps) {
  position <- 0

  for(step in 1:max_steps) {
    move <- ifelse(runif(1) < p, 1, -1)
    position <- position + move

    if(position == target) return(TRUE)

    if(abs(position) > target + 100) return(FALSE)
  }

  return(FALSE)
}

results <- replicate(n_simulations,
                    simulate_single_journey(p, target, max_steps))

probability <- mean(results)

barplot(probability,
        names.arg = "Probability of reaching station 2",
        col = "skyblue",
        main = paste("Probability of Reaching Station 2\n(p=0.3,",
                    n_simulations, "Simulations)"),
        ylim = c(0, 1),
        ylab = "Probability")

text(x = 1,
     y = probability + 0.02,
     labels = sprintf("%.3f", probability))

cat("\nProbability of reaching station 2:", sprintf("%.3f", probability))
```

تأثیر p=0.55:

اما این برتری کم (فقط 0.55) باعث می‌شود حرکت به کندی به سمت جلو باشد

هر چه ایستگاه دورتر باشد، تغییرپذیری در زمان رسیدن بیشتر می‌شود
این به دلیل تجمع عدم قطعیت در هر گام است


رشد غیرخطی زمان:

برای ایستگاه‌های نزدیک، رشد زمان تقریباً خطی است
اما برای ایستگاه‌های دورتر، رشد سریع‌تر می‌شود
این به دلیل احتمال بیشتر برگشت‌های طولانی است

```r

set.seed(123)
n_simulations <- 10000
p <- 0.55
max_k <- 50
max_steps <- 1000

simulate_journey <- function(target_k, p) {
  position <- 0
  steps <- 0

  while(steps < max_steps) {
    move <- ifelse(runif(1) < p, 1, -1)
    position <- position + move
    steps <- steps + 1

    if(position == target_k) {
      return(steps)
    }
  }

  return(max_steps)
}
avg_times <- numeric(max_k)

for(k in 1:max_k) {
  cat("\rSimulating for station", k, "of", max_k, "...")
  times <- replicate(n_simulations, simulate_journey(k, p))
  avg_times[k] <- mean(times[times < max_steps])
}

plot(1:max_k, avg_times,
     type = "l",
     col = "blue",
     lwd = 2,
     xlab = "Station Number (k)",
     ylab = "Average Steps to Reach",
     main = paste("Average Steps to Reach Each Station\n(p =", p, ",",
                 n_simulations, "simulations)"))

points(1:max_k, avg_times, col = "red", pch = 16)

fit <- lm(avg_times ~ poly(1:max_k, 2))
lines(1:max_k, predict(fit), col = "green", lty = 2)

legend("topleft",
       legend = c("Actual Data", "Quadratic Fit"),
       col = c("blue", "green"),
       lty = c(1, 2),
       pch = c(16, NA))

cat("\n\nSummary Statistics:\n")
cat("Min average steps:", min(avg_times), "for station", which.min(avg_times), "\n")
cat("Max average steps:", max(avg_times), "for station", which.max(avg_times), "\n")
```

برای موقعیت نهایی داریم:

E[Zn] = E[Z0] + E[X1 + X2 + ... + Xn] = 100 + n×0 = 100
Var[Zn] = Var[Z0] + Var[X1 + X2 + ... + Xn] = 0 + n×1 = n


با n = 10:

E[Z10] = 100
Var[Z10] = 10
σ = √10 ≈ 3.16


بنابراین Z10 دارای توزیع نرمال با:
Z10 ~ N(100, 10)
ما می‌خواهیم P(Z10 > 105) را محاسبه کنیم.
می‌توانیم این را به یک متغیر نرمال استاندارد تبدیل کنیم:
P(Z10 > 105) = P((Z10 - 100)/√10 > (105 - 100)/√10)
= P(Z > 5/√10)
= P(Z > 1.58)
برای متغیر نرمال استاندارد:
P(Z > 1.58) = 1 - Φ(1.58) ≈ 0.057

بنابراین، احتمال اینکه بعد از 10 حرکت، موقعیت علیرضا بیشتر از 105 باشد حدود 0.057 یا 5.7% است.

با توجه به مورد الف، موقعیت مکانی تاکسی را به کمک شبیه‌سازی تعیین کنید. فرض کنید $X_i \sim N(0, 1)$. حرکت تاکسی را $N = 10000$ بار شبیه‌سازی کنید و برای $n = 10, 20, 50, 100, 200$ تعداد دفعاتی که موقعیت نهایی تاکسی جلوتر از نقطه‌ی 105 است را گزارش کنید.


</font>

```r

set.seed(123)
N <- 10000
z0 <- 100
target <- 105
n_values <- c(10, 20, 50, 100, 200)

calculate_probabilities <- function(n) {

  movements <- matrix(rnorm(N * n, mean = 0, sd = 1),
                     nrow = N,
                     ncol = n)

  final_positions <- z0 + rowSums(movements)

  prob_empirical <- mean(final_positions > target)

  prob_theoretical <- 1 - pnorm(target, mean = z0, sd = sqrt(n))

  return(c(prob_empirical, prob_theoretical))
}

results <- matrix(NA, nrow = length(n_values), ncol = 4)
colnames(results) <- c("n", "Count", "Empirical Prob", "Theoretical Prob")
results[, 1] <- n_values

for(i in 1:length(n_values)) {
  probs <- calculate_probabilities(n_values[i])
  results[i, 2] <- round(probs[1] * N)
  results[i, 3] <- round(probs[1], 4)
  results[i, 4] <- round(probs[2], 4)
}

print(as.data.frame(results))

par(mfrow = c(1, 2))

plot(n_values, results[, 2],
     type = "b",
     col = "blue",
     xlab = "Number of Steps (n)",
     ylab = "Count of Positions > 105",
     main = "Count vs Number of Steps",
     pch = 16)

plot(n_values, results[, 3],
     type = "b",
     col = "red",
     xlab = "Number of Steps (n)",
     ylab = "Probability",
     main = "Empirical vs Theoretical Probability",
     pch = 16)
lines(n_values, results[, 4],
      type = "b",
      col = "green",
      pch = 17)
legend("topleft",
       legend = c("Empirical", "Theoretical"),
       col = c("red", "green"),
       pch = c(16, 17))
```

```r

set.seed(123)
n_values <- c(10, 100, 1000, 10000, 100000)

possible_moves <- matrix(c(1,0, -1,0, 0,1, 0,-1), ncol=2, byrow=TRUE)

simulate_path <- function(n_steps) {
  positions <- matrix(0, nrow=n_steps+1, ncol=2)

  for(i in 1:n_steps) {
    move_idx <- sample(1:4, 1)
    move <- possible_moves[move_idx,]

    positions[i+1,] <- positions[i,] + move
  }

  return(positions)
}

par(mfrow=c(2,3))

for(n in n_values) {
  path <- simulate_path(n)

  plot(path[,1], path[,2],
       type="l",
       col="blue",
       main=paste("Random Walk (n =", n, ")"),
       xlab="X",
       ylab="Y",
       asp=1)

  points(path[1,1], path[1,2], col="green", pch=16)
  points(path[n+1,1], path[n+1,2], col="red", pch=16)

  legend("topright",
         legend=c("Start", "End", "Path"),
         col=c("green", "red", "blue"),
         pch=c(16, 16, NA),
         lty=c(NA, NA, 1),
         cex=0.8)
}

stats <- data.frame(
  n_steps = n_values,
  final_distance = numeric(length(n_values))
)

for(i in 1:length(n_values)) {
  path <- simulate_path(n_values[i])
  final_pos <- path[n_values[i]+1,]
  stats$final_distance[i] <- sqrt(sum(final_pos^2))
}

print(stats)
```

   $X_i \sim \mathcal{N} \left( \begin{bmatrix} \mu_1 \\ \mu_2 \end{bmatrix}, \begin{bmatrix} \sigma_1^2 & 0 \\ 0 & \sigma_2^2 \end{bmatrix} \right)$

```r

library(MASS)

set.seed(123)
n_values <- c(10, 1000, 10000, 100000)
mu1_values <- c(-10, -5, 0, 5, 10)
sigma_values <- c(1, 5, 10, 50, 100)

simulate_path <- function(n_steps, mu1, sigma) {
  positions <- matrix(0, nrow=n_steps+1, ncol=2)

  cov_matrix <- matrix(c(sigma^2, 0, 0, sigma^2), nrow=2)

  moves <- mvrnorm(n_steps, mu=c(mu1, 0), Sigma=cov_matrix)

  positions[-1,] <- t(apply(moves, 2, cumsum))

  return(positions)
}

n <- 1000
par(mfrow=c(length(mu1_values), length(sigma_values)))
par(mar=c(2,2,2,1))

for(mu1 in mu1_values) {
  for(sigma in sigma_values) {
    path <- simulate_path(n, mu1, sigma)

    plot(path[,1], path[,2],
         type="l",
         col="blue",
         main=sprintf("n=%d, μ₁=%d, σ=%d", n, mu1, sigma),
         xlab="X",
         ylab="Y",
         cex.main=0.8)

    points(path[1,1], path[1,2], col="green", pch=16, cex=0.8)
    points(path[n+1,1], path[n+1,2], col="red", pch=16, cex=0.8)

    abline(h=0, v=0, col="gray", lty=2)
  }
}

cat("\nStatistics:\n")
for(mu1 in mu1_values[1:3]) {
  for(sigma in sigma_values[1:3]) {
    path <- simulate_path(n, mu1, sigma)
    final_pos <- path[n+1,]
    distance <- sqrt(sum(final_pos^2))
    cat(sprintf("μ₁=%d, σ=%d: Final distance = %.2f\n",
                mu1, sigma, distance))
  }
}
```

```r
N <- 1000
n <- 1000
p_values <- c(0.1, 0.3, 0.5, 0.7, 0.9)

results <- matrix(0, nrow = length(p_values), ncol = 2)
colnames(results) <- c("p", "return_probability")
results[, 1] <- p_values

for(p_idx in 1:length(p_values)) {
    p <- p_values[p_idx]
    returns <- 0

    for(sim in 1:N) {
        position <- 0
        returned_to_origin <- FALSE

        for(step in 1:n) {
            move <- ifelse(runif(1) < p, 1, -1)
            position <- position + move

            if(position == 0) {
                returned_to_origin <- TRUE
                break
            }
        }

        if(returned_to_origin) {
            returns <- returns + 1
        }
    }

    results[p_idx, 2] <- returns / N
}

print("احتمال بازگشت به مبدا برای مقادیر مختلف p:")
print(results)
```
